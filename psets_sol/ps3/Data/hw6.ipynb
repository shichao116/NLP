```python
twts = []
f = open('oct27.train')
```

```python
twt = ''
for line in f:
    if len(line.split()) > 0 and line != '' and line != ' ':
        twt = twt + line.split()[0] +' '
    else:
        twts.append(twt)
        twt = ''
    
```
